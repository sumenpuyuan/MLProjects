```python
from numpy import *
import cv2
import matplotlib.pyplot as plt
import numpy as np
def loadDataSet(fileName):
    with open(fileName) as f1:
        content = f1.readlines()
        lines = [list(map(float,line.strip("\n").split("\t"))) for line in content]
    return mat(lines)

def pca1(dataMat,topNfeat = 9999):
    meanVal = mean(dataMat,axis=0)
    meanRemoved = dataMat - meanVal
    covMat = cov(meanRemoved,rowvar=0)#协方差矩阵
    eigVals,eigVects = linalg.eig(mat(covMat)) #特征值和特征向量
    eigInd = argsort(eigVals)[::-1][0:topNfeat] #从大到小排序
    redEigVects = eigVects[:,eigInd] #
    lowData = meanRemoved * redEigVects #降维后的数据
    reconMat = (lowData * redEigVects.T) + meanVal #重构后的数据
    return lowData,reconMat
def pca(dataMat,topNfeat=999999):
    meanVals=mean(dataMat,axis=0)
    meanRemoved=dataMat-meanVals
    covMat=cov(meanRemoved,rowvar=0)
    eigVals,eigVects=linalg.eig(mat(covMat))
    eigValInd=argsort(eigVals)
    eigValInd=eigValInd[:-(topNfeat+1):-1]
    redEigVects=eigVects[:,eigValInd]
    lowDDataMat=meanRemoved*redEigVects
    reconMat=(lowDDataMat*redEigVects.T)+meanVals
    return lowDDataMat,reconMat
#读取文件夹下的所有图片 生成一个矩阵
def generateMat():
    lastImgNum = []
    for i in range(1,121):
       
        for j in range(1,15):
            imgNum = []
            path='AR/'+str(i).zfill(3)+'-'+str(j).zfill(2)+'.bmp'
            print('now path is',path)
            img=cv2.imread(path)
            for m in range(0, 50):
                for n in range(0, 40):
                    imgNum.append(img[m][n][0])
            lastImgNum.append(imgNum)
    return mat(lastImgNum)
lastImgNum=generateMat()

```

```python
#显示原始的第一张图片
img=cv2.imread('AR/001-01.bmp')
plt.imshow(img)
plt.show()
```

```python
#展示降维重建后的第一张图片
def showFirstImg(reconMat):
    test=reconMat[0].reshape(50,40)
    test=test.tolist()
    #print(int(test[0][0].real))
    #全部转换成整数
    for i in range(50):
        for j in range(40):
            test[i][j]=test[i][j].real
    plt.imshow(test,cmap='gray')
    plt.show()
#计算重建误差 传入两个矩阵
def calcRebuildError(reconMat,lastImgNum):
    #原始数据在lastImgNum里面 重建后的数据在reconMat里面
    reconList=reconMat.tolist()
    lastImgList=lastImgNum.tolist()
    #计算平方误差
    #存储总的图片数量
    imgNum=shape(lastImgList)[0]
    sum=0
    for i in range(shape(lastImgList)[0]):
        #print('process',i)
        #print('sum is',sum)
        if i%100 == 0:
            print('proecess ',i,' sum is ',sum)
        for j in range(shape(lastImgList)[1]):
            #print(lastImgList[i][j]-reconList[i][j])
            #input()
            sum=sum+(lastImgList[i][j]-reconList[i][j])**2
        sum=sum/j
    sum=sum/imgNum
    #print(sum)
    return sum
```

# 测试降成100维的误差 #

```python
lowDMat,reconMat=pca(lastImgNum,100)
print(calcRebuildError(reconMat,lastImgNum))
```

# 100维特征重建图像 #

```python
showFirstImg(reconMat)
```

# 512维特征降维 #

```python
lowDMat,reconMat=pca(lastImgNum,512)
print(calcRebuildError(reconMat,lastImgNum))
```
